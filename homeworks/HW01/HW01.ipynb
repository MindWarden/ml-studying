```python
# 1. Импорт нумпай
import numpy as np

# 2. Создание массивов
a = np.array([1, 2, 3, 4, 5])              # из списка
b = np.array([[1, 2], [3, 4]])             # 2д-массив
z = np.zeros((2, 3))                       # нули 2×3
o = np.ones((2, 3))                        # единицы 2×3
r = np.arange(0, 10, 2)                    # 0,2,4,6,8

# 3. Атрибуты
print("Форма b:", b.shape)    # (2, 2)
print("Тип b:", b.dtype)      # int64

# 4. Индексы и срезы
print("Элемент [1,0]:", b[1,0])    # 3
print("Первая строка:", b[0])      # [1 2]
print("Первый столбец:", b[:,0])   # [1 3]

# 5. Мат. операции
print("a + a   =", a + a)          # [2 4 6 8 10]
print("a * 10 =", a * 10)          # [10 20 30 40 50]
print("Среднее a:", a.mean())      # 3.0
print("Сумма b:", b.sum())         # 10
```
